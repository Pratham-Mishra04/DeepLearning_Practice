```python
input_sequences = []

for title in data['project_name'].values:
    for i in range(1,len(title)):
        input_sequences.append(title[:i+1])
```

```python
input_sequences
```

```python
input_df = pd.DataFrame(input_sequences, columns=['seq'])
```

```python
def get_substr_max(seq):
    matching_titles = data[data['project_name'].str.startswith(seq[:len(seq)-1])] # cause later it last letter will be set as Y
    max_score=0
    for i in range (0, len(matching_titles)):
        if matching_titles.iloc[i]['score']>max_score:
            max_score=matching_titles.iloc[i]['score']

    return max_score
```

```python
input_df['max_substr_score']=input_df['seq'].apply(get_substr_max)
```

```python
input_df
```

```python
last_input_size = 3
last_input_score = 0

def get_score_by_prefix(input_prefix):
    global last_input_size
    global last_input_score

    if len(input_prefix) < last_input_size:
        matching_titles = data[data['project_name'].str.startswith(input_prefix) & ~data['project_name'].apply(lambda x: x in similar_covered_list)]

        if not matching_titles.empty:
            similar_covered_list.append(matching_titles.iloc[0]['project_name'])
            last_input_score = matching_titles.iloc[0]['score']

    else:
        matching_titles = data[(data['project_name'].str.startswith(input_prefix)) & (data['score'] == last_input_score)]

    last_input_size = len(input_prefix)

    if not matching_titles.empty:
        return matching_titles.iloc[0]['score']
    else:
        return 0

scores = []
similar_covered_list = []

for i in range(len(input_df)):
    s = input_sequences[i]
    scores.append(get_score_by_prefix(s))
```

```python
input_df = pd.concat([input_df, pd.DataFrame(scores, columns=['score'])], axis=1)
```

```python
input_df = input_df[input_df['score']==input_df['max_substr_score']]
```

```python
input_df
```

```python
alphabetes = 'abcdefghijklmnopqrstuvwxyz'
special_characters = ' '

char_dict = {}

letter_to_int = {letter: i+1 for i, letter in enumerate(alphabetes+special_characters)}

for i, letter in enumerate(alphabetes+special_characters):
    char_dict[i+1] = letter

def get_tokens(title):
    tokens = [letter_to_int.get(letter, 0) for letter in title.lower()]
    return tokens
```

```python
tokenized_input_sequences = [get_tokens(seq) for seq in input_df['seq'].values]
```

```python
tokenized_input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(tokenized_input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
y_df=pd.DataFrame(y, columns=['y'])
```

```python
y_df = pd.concat([y_df, pd.get_dummies(y_df['y'])], axis=1)
```

```python
y_df.drop('y', axis=1, inplace=True)
```

```python
y_df
```

```python
for i in range (1, len(alphabetes)+len(special_characters)+1):
    def convert_bool(x):
        if x:
            return 1
        else:
            return 0

    y_df[i]=y_df[i].apply(convert_bool)
```

```python
y = y_df
```

```python
X.shape
```

```python
y.shape
```

```python
from keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Input, concatenate
```

```python
max_characters = len(alphabetes)+len(special_characters)+1

model = Sequential()

model.add(Embedding(max_characters, 100, input_length=max_len-1))
model.add(LSTM(400))
model.add(Dense(32, activation='relu'))
model.add(Dense(max_characters-1, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X, y, epochs=100)
```

```python
text = "a"

predictions = []
 
for i in range(10):
    # tokenize
    token_text = get_tokens(text)
    # padding
    padded_token_text = pad_sequences([token_text], maxlen=max_len-1, padding='pre')
    # predict
    pos = np.argmax(model.predict(padded_token_text))
    
    text = text + char_dict[pos+1]
    if text.strip() not in predictions:
        predictions.append(text)

for p in predictions:
    print(p)
```
